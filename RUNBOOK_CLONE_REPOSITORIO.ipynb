# Runbook: clonar e posicionar o repositorio monorepo-ai-llm

Este notebook documenta e executa, fase por fase, o processo usado para clonar o repositorio:

`https://github.com/marreiro23/monorepo-ai-llm.git`

Destino final esperado:

`/mnt/repositorio/workdir/repostorios/monorepo/monorepo-ai-llm`


## Fase 1: validar o diretorio inicial

Confirme onde a sessao esta executando e veja se a pasta temporaria esta vazia.

```python
%%bash
pwd
ls
```

## Fase 2: clonar o repositorio

Baixe o repositorio do GitHub para a pasta temporaria. Se houver erro de DNS ou rede, execute novamente em um ambiente com acesso ao GitHub liberado.

```python
%%bash
cd /home/daniel/monorepo-template
git clone https://github.com/marreiro23/monorepo-ai-llm.git
```

## Fase 3: conferir o clone

Entre no repositorio clonado, confira o status Git e liste os arquivos principais. O `git status --short` deve sair vazio quando tudo estiver limpo.

```python
%%bash
cd /home/daniel/monorepo-template/monorepo-ai-llm
git status --short
ls
```

## Fase 4: validar o destino definitivo

Confirme que o diretorio `/mnt/repositorio/workdir/repostorios/monorepo` existe e esta pronto para receber o repositorio.

```python
%%bash
ls /mnt/repositorio/workdir/repostorios
ls /mnt/repositorio/workdir/repostorios/monorepo
```

## Fase 5: mover para o destino correto

Move o clone da pasta temporaria do perfil para o caminho definitivo em `/mnt`.

```python
%%bash
mv /home/daniel/monorepo-template/monorepo-ai-llm \
  /mnt/repositorio/workdir/repostorios/monorepo/monorepo-ai-llm
```

## Fase 6: tratar arquivos Zone.Identifier, se necessario

Alguns arquivos vindos de ambientes Windows podem ter `:Zone.Identifier` no nome. O filesystem de `/mnt` pode recusar esses nomes com erro `Invalid argument`.

Se isso acontecer, localize os arquivos que sobraram e marque esses caminhos no indice Git como `skip-worktree`.

```python
%%bash
find /home/daniel/monorepo-template/monorepo-ai-llm -name '*Zone.Identifier' -print
```

```python
%%bash
cd /mnt/repositorio/workdir/repostorios/monorepo/monorepo-ai-llm
git update-index --skip-worktree -- \
  'apps/users-api/src/notebooks_01_nest_monorepo_bootstrap_exec.ipynb:Zone.Identifier' \
  'apps/users-api/src/notebooks_nest-monorepo-guidelines.ipynb:Zone.Identifier' \
  'spec/docs_architecture_project-structure-blueprint (1).md:Zone.Identifier' \
  'spec/docs_architecture_project-structure-blueprint.md:Zone.Identifier' \
  'spec/docs_naming-conventions-plan.md:Zone.Identifier' \
  'spec/fri_may_01_2026_node_version_manager_nvm_e_uma.zip:Zone.Identifier' \
  'spec/notebooks_01_conventions_and_bootstrap.ipynb:Zone.Identifier' \
  'spec/notebooks_01_nest_monorepo_bootstrap_exec (1).ipynb:Zone.Identifier' \
  'spec/notebooks_02_validate_current_config_state (1).ipynb:Zone.Identifier' \
  'spec/spec_spec-microservice-packaging-and-boundaries.md:Zone.Identifier'
git status --short
```

## Fase 7: remover a copia temporaria

Depois de confirmar que o repositorio esta no destino, remova a copia antiga da pasta do perfil.

```python
%%bash
ls /mnt/repositorio/workdir/repostorios/monorepo
rm -rf /home/daniel/monorepo-template/monorepo-ai-llm
```

## Fase 8: validacao final

Confirme que a pasta temporaria ficou vazia, que o repositorio existe em `/mnt`, e que o Git esta limpo.

```python
%%bash
ls /home/daniel/monorepo-template
ls /mnt/repositorio/workdir/repostorios/monorepo
cd /mnt/repositorio/workdir/repostorios/monorepo/monorepo-ai-llm
git status --short
```

## Observacao importante

Este notebook contem comandos destrutivos, especialmente `rm -rf`. Antes de executar a fase de limpeza, confirme sempre que o repositorio ja esta no destino correto.
