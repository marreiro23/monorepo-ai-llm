```python
import subprocess

def run(cmd):
    print(f'$ {cmd}')
    result = subprocess.run(cmd, shell=True, text=True, capture_output=True)
    print(result.stdout)
    if result.returncode != 0:
        print(result.stderr)
    return result.returncode

run('docker stats --no-stream')
```

```python
run('docker compose -f docker-compose.remote.yml logs --tail 200 monorepo-ai-llm users-api llm-ops-api sharepoint-api sync-api')
```

```python
run('docker inspect monorepo-ai-llm --format "{{.HostConfig.RestartPolicy.Name}}"')
run('docker inspect users-api --format "{{.HostConfig.Memory}}"')
run('docker inspect users-api --format "{{.HostConfig.NanoCpus}}"')
```
