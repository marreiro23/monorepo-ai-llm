```python
import subprocess\n
\n
def run(cmd):\n
    print(f'\n$ {cmd}')\n
    result = subprocess.run(cmd, shell=True, text=True, capture_output=True)\n
    print(result.stdout)\n
    if result.returncode != 0:\n
        print(result.stderr)\n
    return result.returncode\n
\n
run('docker stats --no-stream')
```

```python
run('docker compose -f docker-compose.remote.yml logs --tail 200 monorepo-ai-llm users-api llm-ops-api sharepoint-api sync-api')
```

```python
run('docker inspect monorepo-ai-llm --format "{{.HostConfig.RestartPolicy.Name}}"')\n
run('docker inspect users-api --format "{{.HostConfig.Memory}}"')\n
run('docker inspect users-api --format "{{.HostConfig.NanoCpus}}"')
```
