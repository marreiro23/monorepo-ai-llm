# Fase 8 - Governanca de Boundaries e CI\n
Objetivo: preparar verificacoes para impedir regressao arquitetural e reforcar isolamento por dominio.

## Escopo da fase\n
- Verificar imports cruzados proibidos\n
- Validar artefato minimo por servico\n
- Definir criterios de falha no pipeline

```python
import subprocess

def run(cmd):
    print(f'$ {cmd}')
    result = subprocess.run(cmd, shell=True, text=True, capture_output=True)
    print(result.stdout)
    if result.returncode != 0:
        print(result.stderr)
    return result.returncode
```

```python
# Revisao rapida da estrutura por app\n
run('find apps -maxdepth 2 -type d | sort')
```

```python
# Busca padroes de import entre dominios para auditoria inicial\n
run("rg -n \"from '../|from '..\" apps || true")
```

```python
# Inspecao minima das imagens para validar artefato por dominio\n
run('docker images | rg monorepo-ai-llm || true')
```
