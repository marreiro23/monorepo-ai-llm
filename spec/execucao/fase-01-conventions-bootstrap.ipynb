````python
# Exemplo: gerador simples de nome de migration (YYYYMMDDHHMM__desc)
from datetime import datetime
import re

def migration_name(desc: str, dt: datetime | None = None) -> str:
    dt = dt or datetime.now()
    safe = desc.strip().lower()
    safe = re.sub(r"[^a-z0-9]+", "_", safe)
    safe = re.sub(r"_+", "_", safe).strip("_")
    return dt.strftime("%Y%m%d%H%M") + "__" + safe + ".sql"

migration_name("Create users table")
````
