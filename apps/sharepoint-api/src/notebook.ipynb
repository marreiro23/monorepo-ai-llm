# Toolkit de Busca e Manipulação com rg (ripgrep)

Este notebook contém exemplos práticos para:
- Buscar arquivos
- Encontrar strings
- Contar ocorrências
- Exportar resultados
- Substituir valores com sed
- Analisar logs e variáveis de ambiente


## 1. Encontrar arquivos por nome

```bash
!rg --files -g "app.module.ts"
```

## 2. Buscar string com linha e coluna

```bash
!rg -n --column "process\.env"
```

## 3. Buscar ignorando case

```bash
!rg -n -i "apikey"
```

## 4. Contar ocorrências totais

```bash
!rg -o "process\.env" | wc -l
```

## 5. Exportar resultado para TXT

```bash
!rg -n --column "process\.env" > resultado.txt
```

## 6. Extrair variáveis únicas

```bash
!rg -o "process\.env\.[A-Z0-9_]+" | sort -u
```

## 7. Gerar .env.example automaticamente

```bash
!rg -o "process\.env\.[A-Z0-9_]+" \
| sed 's/process\.env\.//' \
| sort -u \
| awk '{print $0"=PLACEHOLDER"}' > .env.example
```

## 8. Substituir valores com sed (com backup)

```bash
!sed -i.bak 's/OLD_VALUE/NEW_VALUE/g' arquivo.txt
```

## 9. Substituir process.env por config

```bash
!rg -l "process\.env\." \
| xargs sed -i.bak 's/process\.env\./config./g'
```

## 10. Análise de logs

```bash
!rg -n -i -C 3 "error|failed|timeout|exception" logs/
```

## 11. Sanitizar logs (remover secrets)

```bash
!sed -i.bak -E 's/(token|password|secret)=\S+/\1=REDACTED/gI' logs/*.log
```

## 12. Buscar em URL

```bash
!curl -s https://example.com/file.txt | rg -n "error"
```
