# NestJS CLI Monorepo — Regras mínimas (apps por domínio)

## Pergunta: `app.module.ts` tem que existir em todas as apps e no app root?

**Sim.** Em NestJS, cada aplicação (cada projeto em `apps/<nome>`) precisa do seu próprio `AppModule`.

- `apps/users-api/src/app.module.ts`
- `apps/llm-ops-api/src/app.module.ts`
- `apps/sharepoint-api/src/app.module.ts`
- `apps/sync-api/src/app.module.ts`
- `apps/monorepo-ai-llm/src/app.module.ts` (app root do monorepo; pode ser placeholder)

## Os valores precisam ser iguais?

**Não.** Cada `AppModule` deve conter apenas os controllers/providers do domínio correspondente.
O que deve ser comum é a disciplina: **mínimo necessário**, e evitar acoplamento entre apps.


## Template: HealthController (recomendado em todas as apps)

Crie `apps/<app>/src/health.controller.ts`:


````python
health_controller = '''\
import { Controller, Get } from '@nestjs/common';

@Controller('health')
export class HealthController {
  @Get()
  getHealth() {
    return { status: 'ok' };
  }
}
'''
print(health_controller)
````

## Template: AppModule mínimo com Health

```ts
import { Module } from '@nestjs/common';
import { AppController } from './app.controller';
import { AppService } from './app.service';
import { HealthController } from './health.controller';

@Module({
  imports: [],
  controllers: [AppController, HealthController],
  providers: [AppService],
})
export class AppModule {}
```
